```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')
data.head
```

```python
data.describe
```

```python
df=pd.DataFrame(data)
```

```python
y=df[' Estimated Unemployment Rate (%)']
x=df['Region']
```

```python
plt_1 = plt.figure(figsize=(40, 15))
plt.title('Umemployment Rate', fontweight='bold' ,fontsize=20)

plt.xlabel("States",fontweight='bold',fontsize=20)
plt.ylabel("Estimated Unemployment rate",fontweight='bold',fontsize=20)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.bar(x,y, color='r',align='center')
```
